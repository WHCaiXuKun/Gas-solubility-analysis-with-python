```python
import pandas
```

```python
df = pandas.read_excel("demo.xlsx")
df
```

```python
df = df.loc[7:, ["Unnamed: 1", "Unnamed: 2", "Unnamed: 3", "Unnamed: 4"]]
df
```

```python
df.columns=["温度T[K]	", "窒素の溶解度x[モル分率]", "酸素の溶解度x[モル分率]", "水素の溶解度x[モル分率]"]
df.index=list(range(1, 13))
df
```

```python
from sympy import *
```

```python
A = -48.1611
B = 55.2845
C = 16.8893
```

```python
for i in df["温度T[K]\t"]:
    a = exp(A+B/(i/100) + C*log(i/100))
    b = Float(a, 5)
    print(b)
    df["水素の溶解度x[モル分率]"] = b
```

```python
df.to_csv("気体の溶解度.csv")
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager
from scipy.interpolate import make_interp_spline
```

```python
my_font = font_manager.FontProperties(fname="/System/Library/Fonts/ヒラギノ明朝 ProN.ttc")#fontを得る
x = df["温度T[K]\t"]
y_N2 = df["窒素の溶解度x[モル分率]"]
y_O2 = df["酸素の溶解度x[モル分率]"]
y_H2 = df["水素の溶解度x[モル分率]"]
```

```python
x_smooth = np.linspace(x.min(), x.max(), 300)
y_smooth_N2 = make_interp_spline(x, y_N2)(x_smooth)
y_smooth_O2 = make_interp_spline(x, y_O2)(x_smooth)
y_smooth_H2 = make_interp_spline(x, y_H2)(x_smooth)
```

```python
plt.figure(figsize=(20, 8), dpi=80)
plt.title("気体への溶解度", fontproperties=my_font, size=24)
plt.xlabel("温度T[K]", fontproperties=my_font, size=24)
plt.ylabel("気体の溶解度", fontproperties=my_font, size=24)
plt.plot(x_smooth, y_smooth_N2, "--")
plt.plot(x_smooth, y_smooth_O2, ":")
plt.plot(x_smooth, y_smooth_H2)
plt.scatter(x, y_N2,  cmap='rainbow', marker='o', s=200,c=y_N2, alpha=0.4,label="窒素")
plt.scatter(x, y_H2,  cmap='rainbow', marker='v', s=200,c=y_H2, alpha=0.4,label="水素")
plt.scatter(x, y_O2,  cmap='rainbow', marker='s', s=200,c=y_O2, alpha=0.4,label="酸素")
plt.legend(prop=my_font)
plt.savefig("気体の溶解度.png")
plt.show()
```
